```python
import pandas as pd
# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
file_path = '/content/drive/MyDrive/Colab Notebooks/Salary_Data.csv'
# Read CSV file into a pandas DataFrame
df = pd.read_csv(file_path)
# Display the first few rows of the DataFrame
df.head()
```

```python
x=df['YearsExperience']
y=df['Salary']
```

```python
import matplotlib.pyplot as plt



# Plotting the data
plt.scatter(x, y, label='Scatter Plot')

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
X_train = X_train.values.reshape((24, 1))
y_train=y_train.values.reshape((24, 1))
```

```python
from keras.models import Sequential
from keras.layers import Dense

# Create a sequential model
model = Sequential()
# units=1: Single output neuron for regression
# input_dim=1: Assuming your dataset has a Single feature

# Add a single dense layer for linear regression
model.add(Dense(units=1, input_dim=1))

# Compile the model
model.compile(optimizer='Adam', loss='mean_squared_error')
#For small datasets, you might use one batch per epoch.
# Train the model
history=model.fit(X_train, y_train, epochs=100, batch_size=32,validation_data=(X_test, y_test))
# Accessing the training and validation losses over epochs
training_loss = history.history['loss']
validation_loss = history.history['val_loss']
```

```python
# Print the last loss values
print(f"Final Training Loss: {training_loss[-1]}")
print(f"Final Validation Loss: {validation_loss[-1]}")
```

```python
import matplotlib.pyplot as plt

# Assuming you have already trained the model and stored the history

# Plotting the training and validation losses over epochs
plt.plot(training_loss, label='Training Loss')
plt.plot(validation_loss, label='Validation Loss')

# Adding labels and title
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss Over Epochs')

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```

```python
from keras.models import Sequential
from keras.layers import Dense,Input

# Create a sequential model
model = Sequential()

model.add(Input(shape=(1,)))
model.add(Dense(units=1))

# Compile the model
model.compile(optimizer='SGD', loss='mean_squared_error')
#For small datasets, you might use one batch per epoch.
# Train the model
history=model.fit(X_train, y_train, epochs=100, batch_size=32,validation_data=(X_test, y_test))
# Accessing the training and validation losses over epochs
training_loss = history.history['loss']
validation_loss = history.history['val_loss']
```

```python
# Print the last loss values
print(f"Final Training Loss: {training_loss[-1]}")
print(f"Final Validation Loss: {validation_loss[-1]}")
```

```python
import matplotlib.pyplot as plt

# Assuming you have already trained the model and stored the history

# Plotting the training and validation losses over epochs
plt.plot(training_loss, label='Training Loss')
plt.plot(validation_loss, label='Validation Loss')

# Adding labels and title
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss Over Epochs')

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```
